```python
from PIL import Image
```

```python
# KHAI BÁO ĐƯỜNG DẪN CHỨA HÌNH
dir = 'C:/Users/trant/Pictures/Screenshots'
```

```python
image_path = dir + '/1.png'
print(image_path)
```

```python
img01 = Image.open(image_path)
```

```python
display(img01)
```

```python
# in định dạng ảnh
print("Định dạng ảnh: ", img01.format)
```

```python
# in kích cỡ của ảnh
print("Kích thước ảnh: ", img01.size)
```

```python
# convert RGBA về RGB để đổi được file PNG về JPG
rgb_img01 = img01.convert('RGB')
```

```python
# lưu và chuyển đổi hình ảnh
new_img_01_path = dir + '/new_img01.png'
new_img_JPG_01_path = dir + '/new_img01.JPG'
rgb_img01.save(new_img_01_path)
rgb_img01.save(new_img_JPG_01_path)
```

```python
# Đóng tập tin
rgb_img01.close()
```
